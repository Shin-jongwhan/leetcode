```python
from collections import deque

class TreeNode(object):
	def __init__(self, val=0, left=None, right=None):
		self.val = val
		self.left = left
		self.right = right


def generate_tree_node(ls):
	if not ls:
		return None  # 리스트가 비어 있으면 빈 트리 반환

	# 루트 노드 생성
	root = TreeNode(ls[0])
	queue = [root]  # BFS로 노드를 할당하기 위한 큐
	i = 1  # 리스트의 다음 값을 가리킬 인덱스

	while i < len(ls):
		current_node = queue.pop(0)  # 큐에서 노드를 가져옴

		# 왼쪽 자식 노드 할당
		if i < len(ls) and ls[i] is not None:
			current_node.left = TreeNode(ls[i])
			queue.append(current_node.left)
		i += 1

		# 오른쪽 자식 노드 할당
		if i < len(ls) and ls[i] is not None:
			current_node.right = TreeNode(ls[i])
			queue.append(current_node.right)
		i += 1

	return root


def print_tree(root):
	if not root:
		print("Tree is empty.")
		return

	# BFS를 이용한 레벨 순서대로 출력
	queue = deque([root])
	while queue:
		level_size = len(queue)
		level_nodes = []
		for _ in range(level_size):
			node = queue.popleft()
			if node:
				level_nodes.append(str(node.val))
				queue.append(node.left)
				queue.append(node.right)
			else:
				level_nodes.append("None")
		
		# 레벨별로 출력
		print(" | ".join(level_nodes))


def test(root) : 
	# 리스트로 깊이 별로, [left, right]로 묶어서 리턴하기
	# 예를 들어 [[1], [1,2]] 이런 식이다.
	# BFS를 활용하면 좋겠다.
	# deque로 왼쪽부터 오른쪽으로 하나씩 넣는다. depth를 알아야 하니까 (node, depth) 형식으로 넣는다.
	# popleft 하여 다음 값을 확인한다.
	if not root : 
		return []

	depth = 1
	dq = deque([(root, depth)])
	ls = []
	while dq : 
		node, depth = dq.popleft()
		if node.left : 
			dq.append((node.left, depth + 1))
		if node.right : 
			dq.append((node.right, depth + 1))
		if len(ls) == depth : 
			ls[-1].append(node.val)
		else : 
			ls.append([node.val])

	return ls


ls = [1,2,2,3,4,4,3]
#ls = [1,0]
#ls = [1]
ls = [1,2,3,None,4,None,5]
#ls = []
root = generate_tree_node(ls)
#print_tree(root)
output = test(root)
print(output)
```
