```python
from collections import deque

class ListNode(object):
	def __init__(self, x=0, next=None):
		self.val = x
		self.next = None
		

def create_linked_list(ls) : 
	if ls == [] :
		return None
	ln = ListNode(ls[0])
	next = ln		# next는 ln과 동일한 첫 번째 노드를 참조함. 완전히 새로운 변수가 아니기 때문에 ln의 업데이트가 가능.
	for i in range(1, len(ls)) : 
		next.next = ListNode(ls[i])
		next = next.next

	return ln


def print_linked_list(node):
    while node:
        print(node.val, end=" -> ")
        node = node.next
    print("None")


def test(ln) : 
	if not ln : 
		return False
	
	stAddress = set([id(ln)])
	current = ln

	while current : 
		if id(current.next) not in stAddress : 
			stAddress.add(id(current.next))
		else : 
			return True
		current = current.next

	return False


ls = [1,2,3,2,1]
ls = [1,2,2,1]
ls = [1,2,2,1,1]
ln = create_linked_list(ls)

test(ln)
```
