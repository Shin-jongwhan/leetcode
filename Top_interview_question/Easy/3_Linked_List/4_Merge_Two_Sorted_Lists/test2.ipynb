```python
######################################################
# 주어진 조건
# 1. 두 linked list 모두 오름차순으로 되어 있다.
# 2. 0 <= length <= 50이다.
#
# 생각해볼 점
# 1. 두 node가 모두 비어 있으면? -> None으로 됨
# 2. 둘 중 하나가 None이면? -> None을 검사해서 하나만 계속 append
# 3. 둘 다 오름차순으로 되어 있다면, 둘 중 하나의 시작 value부터 순차적으로 추가하면 된다.
#
# TODO
# 1. while 문으로 둘 다 None / falsy가 아니라면(if node1 and node2 : ) 으로 들어간다.
######################################################

class ListNode(object):
	def __init__(self, x=0, next=None):
		self.val = x
		self.next = None
		

def create_linked_list(ls) : 
	if ls == [] :
		return None
	ln = ListNode(ls[0])
	next = ln		# next는 ln과 동일한 첫 번째 노드를 참조함. 완전히 새로운 변수가 아니기 때문에 ln의 업데이트가 가능.
	for i in range(1, len(ls)) : 
		next.next = ListNode(ls[i])
		next = next.next

	return ln


def print_linked_list(node):
    while node:
        print(node.val, end=" -> ")
        node = node.next
    print("None")


def test(ln1, ln2) : 
	if not ln1 or not ln2 :
		return ln1 if ln1 else ln2
	current1 = ln1
	current2 = ln2
	lnConcat = ListNode()
	next = lnConcat
	while current1 or current2 : 
		if current1 and current2 : 
			if current1.val >= current2.val : 
				next.val = current2.val
				current2 = current2.next
			else : 
				next.val = current1.val
				current1 = current1.next
		elif current1 : 
			next.val = current1.val
			current1 = current1.next
		elif current2 : 
			next.val = current2.val
			current2 = current2.next
		next.next = ListNode() if current1 or current2 else None
		next = next.next

	return lnConcat

list1 = [1,2,4]
list2 = [1,3,4]

#list1 = []
#list2 = []

#list1 = [1,2,3]
#list2 = []

#list1 = []
#list2 = [1,2,3]

ln1 = create_linked_list(list1)
ln2 = create_linked_list(list2)

lnConcat = test(ln1, ln2)
print_linked_list(lnConcat)
```
