```python
board = \
[["5","3",".",".","7",".",".",".","."]
,["6",".",".","1","9","5",".",".","."]
,[".","9","8",".",".",".",".","6","."]
,["8",".",".",".","6",".",".",".","3"]
,["4",".",".","8",".","3",".",".","1"]
,["7",".",".",".","2",".",".",".","6"]
,[".","6",".",".",".",".","2","8","."]
,[".",".",".","4","1","9",".",".","5"]
,[".",".",".",".","8",".",".","7","9"]]

"""
board = \
[["8","3",".",".","7",".",".",".","."]
,["6",".",".","1","9","5",".",".","."]
,[".","9","8",".",".",".",".","6","."]
,["8",".",".",".","6",".",".",".","3"]
,["4",".",".","8",".","3",".",".","1"]
,["7",".",".",".","2",".",".",".","6"]
,[".","6",".",".",".",".","2","8","."]
,[".",".",".","4","1","9",".",".","5"]
,[".",".",".",".","8",".",".","7","9"]]
"""

board = \
[["7",".",".",".","4",".",".",".","7"],
 [".",".",".","8","6","5",".",".","."],
 [".","1",".","2",".",".",".",".","."],
 [".",".",".",".",".","9",".",".","."],
 [".",".",".",".","5",".","5",".","."],
 [".",".",".",".",".",".",".",".","."],
 [".",".",".",".",".",".","2",".","."],
 [".",".",".",".",".",".",".",".","."],
 [".",".",".",".",".",".",".",".","."]]

# 스도쿠가 이렇게 시작해도 되는지에 대한 문제. 문제가 있는지 검사한다.
# cols = [[0, 1, 2], [3, 4, 5], [6, 7, 8]]
# rows = [[0, 1, 2], [3, 4, 5], [6, 7, 8]]
# 1. 먼저 col * row에 중복되는 숫자가 있으면 안 된다. 
# 2. 한 row에 중복되는 숫자가 있으면 안 된다.
# 3. 한 col에 중복되는 숫자가 있으면 안 된다.


def validation(ls) : 
	if len(ls) == len(set(ls)) : 
		return True
	else : 
		return False


cols = [[0, 1, 2], [3, 4, 5], [6, 7, 8]]
rows = [[0, 1, 2], [3, 4, 5], [6, 7, 8]]
blCheck = True

# 1. 먼저 col * row에 중복되는 숫자가 있으면 안 된다. 
for col in cols : 
	lsTmp = []
	for row in rows : 
		lsTmp = [board[i][j] for i in col for j in row if board[i][j] != '.']
		#print(lsTmp)
		blCheck = validation(lsTmp)
		if blCheck == False : 
			break
	if blCheck == False : 
		break
print("1", blCheck)

# 2. 한 row에 중복되는 숫자가 있으면 안 된다.
# 3. 한 col에 중복되는 숫자가 있으면 안 된다.
cols = [0, 1, 2, 3, 4, 5, 6, 7, 8]
rows = [0, 1, 2, 3, 4, 5, 6, 7, 8]
lsCol = []
lsRow = []
for i in cols :
	for j in rows : 
		if board[i][j] != '.' : 
			lsRow.append(board[i][j])
		if board[j][i] != '.' : 
			lsCol.append(board[j][i])
	blCheck = validation(lsRow)
	if blCheck == False : 
		break
	blCheck = validation(lsCol)
	if blCheck == False : 
		break
	lsRow = []
	lsCol = []

print("2", blCheck)
```
